```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Create an array from [0,1] in 100 steps
x = np.arange(0,1,0.01)
```

```python
#Define functions that return sin(x) and cos(x)
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
#Create a miltipanel plot (1 row, 2 columns)
f, axarr = plt.subplots(1,2)

#sin(x) vs x on the left
axarr[0].plot(x,sin(x))

#cos(x) vs x on the right
axarr[1].plot(x,cos(x))

#Label sin(x) and cos(x)
axarr[0].set_title('sin(x)')
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[1].set_title('cos(x)')
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')

f.subplots_adjust(wspace=0.4)

#Save as pdf
plt.savefig('journal-prompt-8.pdf',bbox_inches="tight")
```
