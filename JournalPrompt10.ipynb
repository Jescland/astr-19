```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#use numpy to pull 1000 random numbers distributed uniformly between [0,1]
n = 1000
x = np.random.poisson(2,n)

#histogram the numbers into 100 bins
width = 1
histmin = np.floor(min(x))
histmax = np.ceil(max(x))
bins = np.arange(histmin,histmax,width)

#plot the histogram
plt.hist(x,bins=bins,alpha=0.5,edgecolor="black")

#label axes
plt.xlabel('x')
plt.ylabel('N per bin')

#save as pdf
plt.savefig('journal-prompt-10.pdf',bbox_inches='tight')
```
