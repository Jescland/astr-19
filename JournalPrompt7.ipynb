```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Write cells that create array x ranging from [0,1] in 100 steps
x = np.arange(0,1,0.01)
#Define a function that returns exp(x)
def exp(x):
    return np.exp(x)
```

```python
#Use function to set y=exp(x)
y = exp(x)

#plot x vs y
plt.plot(x,y)

#label x axis as "Time [milliseconds]'
plt.xlabel('Time [milliseconds]')
#label y axis as "Awesomeness"
plt.ylabel('Awesomeness')
#Save figure as a PDF
plt.savefig('journal-prompt-7.pdf',bbox_inches="tight")
```
