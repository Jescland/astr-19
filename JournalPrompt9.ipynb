```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#use numpy to pull 1000 random numbers distributed uniformly between [0,1]
n = 1000
x = np.random.randn(n)

#histogram the numbers into 100 bins
bins = 100

#plot the histogram
plt.hist(x,bins=bins,alpha=0.5,edgecolor="black")

#label axes
plt.xlabel('x')
plt.ylabel('N per bin')

#save as pdf
plt.savefig('journal-prompt-9.pdf',bbox_inches='tight')
```
